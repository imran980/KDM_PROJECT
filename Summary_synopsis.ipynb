```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks_KDM/CSEE5590-KDM/kdm_project/nsf_funding1.csv')
```

```python
df['Synopsis'].replace('NaN', '')
df.drop_duplicates(subset=['Synopsis'],inplace=True)  #dropping duplicates
df['Synopsis'].fillna('', inplace=True)   #dropping na
stop_words = set(stopwords.words('english')) 
df['Synopsis'].apply(lambda x: x.replace('\n',''))
df['Synopsis'].apply(lambda x: x.replace('nan',''))
```

```python
df['Synopsis']
df = df.head(10)
```

```python
df['Synopsis'] = df['Synopsis'].astype(str)
summaries = []
for index, row in df.iterrows():
  syn = df['Synopsis'][index]
  if len(syn) > 10:
    #print("original", len(syn) , "min len" , 0.15 * len(syn) , "max len" , 0.25 * len(syn))
    print(index)  
    print(len(syn))
    summaries.append(summarizer(syn, min_length = int(0.01 * len(syn)), max_length = int(0.1 * len(syn)),do_sample=False)[0]['summary_text'])
    #df['summ'].loc[index] = summary
  else:
    summaries.append(syn)
df['summary'] = summaries
    
```
